```r
library(tidyverse)
```

```r
xcell <- read_tsv('build/xcell-output.tsv')
head(xcell)
```

```r
xcell.T <- xcell %>%
            as.data.frame %>%
            tibble::column_to_rownames('CellType') %>%
            t %>%
            as.data.frame
head(xcell.T)
```

```r
ggplot(xcell.T) + geom_point(mapping=aes(x=`CD4+ T-cells`, y=`CD8+ T-cells`, alpha=MicroenvironmentScore))
```

```r
ggplot(xcell.T) + geom_point(mapping=aes(x=`Th1 cells`, y=`Th2 cells`, alpha=ImmuneScore))
```

```r
ggplot(xcell.T) + geom_point(mapping=aes(x=`Tregs`, y=`CD8+ T-cells`, alpha=ImmuneScore))
```
